```python
from pathlib import Path
from antismash.common import record_processing
from Bio import SeqIO
import pandas as pd
```

```python
# Set input paths
data_dir = Path("../data/APAsp1422609747/Mycocosm/")
fasta = data_dir / "Assembly/Assembled scaffolds (masked)/Aspacri1_AssemblyScaffolds_Repeatmasked.fasta"
taxon = "fungi"
gff = data_dir / 'Annotation/Filtered Models ("best")/Genes/Aspacri1_GeneCatalog_20160412.gff3'
kog = data_dir / 'Annotation/Filtered Models ("best")/Functional Annotations/KOG/Aspacri1_GeneCatalog_proteins_20160412_KOG.tab'
```

```python
# parse with antismash gff parser
records = record_processing.parse_input_sequence(fasta, taxon, gff_file=gff)

# convert back to biopython
records_biopython = [r.to_biopython() for r in records]
```

```python
# write to genbank
SeqIO.write(records_biopython, "my_example_as.gbk", "genbank")
```

```python
kog_tab = pd.read_csv(kog, sep="\t")
```
